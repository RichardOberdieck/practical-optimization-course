# Deviation estimation
First, let us specifically state what we want to do: given a set of 2-D data points $(x,y)$, find the coefficients $a$ and $b$ such that a certain error metric is minimized. The question behind this exercise is to formulate the underlying linear programming problem for linear error metrics.

The regression curve we are going to formulate has the following form: $y = ax + b$. Therefore, given a point $(x_j,y_j)$, the error $z_j$ is given by:
\begin{equation}
z_j = ax_j + b - y_j
\end{equation}

## Initialization and data generation
To do regression, we need some data. Therefore, we generate the following random data set.

```python
import numpy as np
import matplotlib.pyplot as plt
from docplex.mp.model import Model
import math

n = 100
nRange = range(n)
coord = np.random.rand(n,2);

# Let's have a look
plt.plot(coord[:,0], coord[:,1],'r.')
plt.xlabel('x coordinate')
plt.ylabel('y coordinate')

# Define the plotting for post-processing
def plotting(a,b): 
    aVal = mdl.solution.get_var_value(a)
    bVal = mdl.solution.get_var_value(b)
    x = np.linspace(0,1,100)

    plt.plot(coord[:,0], coord[:,1],'r.')
    plt.plot(x,aVal*x + bVal)
    plt.xlabel('x coordinate')
    plt.ylabel('y coordinate');
```

## Least absolute deviations estimation
### Modelling
In the least absolute deviations estimation, we use the following objective function $\sum \limits_{j\in J} \left|z_j\right|$. The idea is that we minimize the linear deviation from the main regression line without penalizing outliers more (as done in higher-norm variations). This means that we want to solve has the following form:
\begin{equation}
\begin{array}{ll}
\underset{a,b,\{z_j, \forall j\}}{\text{minimize}} & \sum \limits_{j} \left|z_j\right| \\
\text{subject to} & z_j = ax_j + b - y_j
\end{array}
\end{equation}

However, we cannot have the absolute value $\left|\cdot\right|$ just like this. Therefore, we have to introduce two auxiliary variables per absolute value formulation:
\begin{align}
z_j &= z_j^+ - z_j^- \\
\left|z_j\right| &= z_j^+ + z_j^-
\end{align}
with $z_j^+,z_j^-\geq 0$. This means, we are actually going to solve the following:
\begin{equation}
\begin{array}{ll}
\underset{a,b,\{z_j, \forall j\}}{\text{minimize}} & \sum \limits_{j} z_j^+ + z_j^- \\
\text{subject to} & z_j^+ - z_j^- = ax_j + b - y_j \\
& z_j^+,z_j^- \geq 0, \hspace{0.3cm} \forall j
\end{array}
\end{equation}

### Problem formulation
Now let's put this into CPLEX:

```python
# Model definition
mdl = Model(name = "Least absolute deviations estimation")
```

#### Variable definition
We start by definiting the variables. Note that Python indexes with 0, i.e. we will have $z_0$ to $z_{99}$:

```python
zPlus = mdl.continuous_var_list(n, lb = 0, name="z+");
zMinus = mdl.continuous_var_list(n, lb = 0, name="z-");
a = mdl.continuous_var(name="a");
b = mdl.continuous_var(name="b");
```

#### Objective function definition

```python
mdl.minimize(mdl.sum(zPlus[j] + zMinus[j] for j in nRange))
```

#### Constraint definition

```python
mdl.add_constraints([zPlus[j] - zMinus[j] == a*coord[j,0] + b - coord[j,1],
                     'Regression %s' % (j)] for j in nRange);
```

### Solving the model and post-processing

```python
mdl.solve()

plotting(a,b)
```

Let's also analyze the values of `zPlus` and `zMinus`:

```python
zPVal = mdl.solution.get_values(zPlus);
zMVal = mdl.solution.get_values(zMinus);

x = np.linspace(0,1,100)
plt.plot(x,zPVal,'r.')
plt.plot(x,np.negative(zMVal),'b.')
plt.xlabel('x coordinate')
plt.ylabel('$z_j$')
```

#### Some comments and ideas
1. We can also leave the equation explicitly in, instead of substituting it:
\begin{equation}
\begin{array}{ll}
\underset{a,b,\{z_j, \forall j\}}{\text{minimize}} & \sum \limits_{j} z_j^+ + z_j^- \\
\text{subject to} & z_j = ax_j + b - y_j \\
& z_j = z^+ - z_j^- \\
& z_j^+,z_j^- \geq 0, \hspace{0.3cm} \forall j
\end{array}
\end{equation}

```python
mdl = Model(name = "Least absolute deviations estimation - test")
z = mdl.continuous_var_list(n, lb = -math.inf, name="z");
zPlus = mdl.continuous_var_list(n, lb = 0, name="z+");
zMinus = mdl.continuous_var_list(n, lb = 0, name="z-");
a = mdl.continuous_var(name="a");
b = mdl.continuous_var(name="b");
mdl.minimize(mdl.sum(zPlus[j] + zMinus[j] for j in nRange))
mdl.add_constraints([z[j] == zPlus[j] - zMinus[j],'z-link %s' % (j)] for j in nRange);
mdl.add_constraints([z[j] == a*coord[j,0] + b - coord[j,1],'Regression %s' % (j)] for j in nRange);
mdl.solve()

plotting(a,b)
```

2. The function `abs` is included in CPLEX, so let's use that instead.

```python
mdl = Model(name = "Least absolute deviations estimation - testAbs")
z = mdl.continuous_var_list(n, lb = -math.inf,name="z");
a = mdl.continuous_var(name="a");
b = mdl.continuous_var(name="b");
mdl.minimize(mdl.sum(mdl.abs(z[j]) for j in nRange))
mdl.add_constraints([z[j] == a*coord[j,0] + b - coord[j,1],'Regression %s' % (j)] for j in nRange);
mdl.solve()

plotting(a,b)
```

## Least maximum deviation estimation
Let's do the other case now:
\begin{equation}
\begin{array}{ll}
\underset{a,b,\{z_j, \forall j\}}{\text{minimize}} & \max \limits_{j} z_j^+ + z_j^- \\
\text{subject to} & z_j^+ - z_j^- = ax_j + b - y_j \\
& z_j^+,z_j^- \geq 0, \hspace{0.3cm} \forall j
\end{array}
\end{equation}
Using the epigraph, we can therefore get:
\begin{equation}
\begin{array}{ll}
\underset{a,b,\{z_j, \forall j\}}{\text{minimize}} & t \\
\text{subject to} & z_j^+ - z_j^- = ax_j + b - y_j \\
& z_j^+ + z_j^- \leq t \\
& z_j^+,z_j^- \geq 0, \hspace{0.3cm} \forall j
\end{array}
\end{equation}
Putting this into CPLEX we get:

```python
mdl = Model(name = "Least maximum deviation estimation")
zPlus = mdl.continuous_var_list(n, lb = 0, name="z+");
zMinus = mdl.continuous_var_list(n, lb = 0, name="z-");
a = mdl.continuous_var(name="a");
b = mdl.continuous_var(name="b");
t = mdl.continuous_var(name="t");
mdl.minimize(t)
mdl.add_constraints([zPlus[j] - zMinus[j] == a*coord[j,0] + b - coord[j,1],'Regression %s' % (j)] for j in nRange);
mdl.add_constraints([zPlus[j] + zMinus[j] <= t, 'Max value %s' % (j)] for j in nRange);
mdl.solve()

plotting(a,b)
```

Let's try the CPLEX reformulation now:

```python
mdl = Model(name = "Least maximum deviation estimation")
z = mdl.continuous_var_list(n, lb = -math.inf,name="z");
a = mdl.continuous_var(name="a");
b = mdl.continuous_var(name="b");
mdl.minimize(mdl.max(mdl.abs(z[j]) for j in nRange))
mdl.add_constraints([z[j] == a*coord[j,0] + b - coord[j,1],'Regression %s' % (j)] for j in nRange);
mdl.solve()

plotting(a,b)
```

## Comparison
As a final point, let's compare the two:

```python
plotting(a,b)

mdl = Model(name = "Least absolute deviations estimation")
z = mdl.continuous_var_list(n, lb = -math.inf,name="z");
a = mdl.continuous_var(name="a");
b = mdl.continuous_var(name="b");
mdl.minimize(mdl.sum(mdl.abs(z[j]) for j in nRange))
mdl.add_constraints([z[j] == a*coord[j,0] + b - coord[j,1],'Regression %s' % (j)] for j in nRange);
mdl.solve()

plotting(a,b)
```
